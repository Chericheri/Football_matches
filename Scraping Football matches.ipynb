```python
#NOW WE ARE GOING TO SCRAP DATA FOR MULTIPLE SEASONS AND TEAMS WITH A FOR LOOP

years = list(range(2025, 2022, -1))
```

```python
years
```

```python
all_matches = []
```

```python
standings_url ="https://fbref.com/en/comps/9/{year}-Premier-League-Stats"
```

```python
import pandas as pd
import time
import requests
from bs4 import BeautifulSoup
from io import StringIO

for year in years:
     # Request data
    data = requests.get(standings_url)
    Soup = BeautifulSoup(data.text, 'html.parser')

    # Check if the table exists
    standings_table_list = Soup.select('table.stats_table')
    
    if standings_table_list:  # Only proceed if table exists
        standings_table = standings_table_list[0]  # Extract the first table

        # Extract links to team pages
        links = [l.get("href") for l in standings_table.find_all('a')]
        links = [l for l in links if '/squads/' in l]
        team_urls = [f"https://fbref.com{l}" for l in links]

        # Get previous season link (for navigating to the prior year's page)
        previous_season = Soup.select("a.prev")
        if previous_season:
            previous_season = previous_season[0].get("href")
            standings_url = f"https://fbref.com{previous_season}"

        # Loop through each team URL
        for team_url in team_urls:
            team_name = team_url.split("/")[-1].replace("-stats", "").replace("-", " ")

            # Request team data
            data = requests.get(team_url)
            matches = pd.read_html(StringIO(data.text), match="Scores & Fixtures")[0]

            # Extract shooting data from the team page
            Soup = BeautifulSoup(data.text, 'html.parser')
            links = [l.get("href") for l in Soup.find_all('a')]
            links = [l for l in links if l and 'all_comps/shooting/' in l]

            if links:  # If there is a shooting data link
                data = requests.get(f"https://fbref.com{links[0]}")
                shooting = pd.read_html(StringIO(data.text), match="Shooting")[0]
                shooting.columns = shooting.columns.droplevel()

                try:
                    # Merge matches with shooting stats on 'Date'
                    team_data = matches.merge(shooting[["Date", "Sh", "SoT", "Dist", "FK", "PK", "PKatt"]], on="Date")
                except ValueError:
                    continue

                # Filter for Premier League games and add team data
                team_data = team_data[team_data["Comp"] == "Premier League"]
                team_data["Season"] = year
                team_data["Team"] = team_name

                # Append the team data to all_matches
                all_matches.append(team_data)

            # Add a sleep interval to avoid hitting the server too quickly
            time.sleep(1)
    else:
        print(f"⚠️ No table found for {year}. Skipping...")
```

```python
len(all_matches)
```

```python
match_df = pd.concat(all_matches, ignore_index=True)
```

```python
match_df.columns = [c.lower() for c in match_df.columns]
```
